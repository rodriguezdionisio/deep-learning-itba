# Fetch weather data

Traigo los datasets de clima que fueron extraídos de Visual Crossing y almacenados como CSV en Google Drive (https://www.visualcrossing.com).

Son dos datasets que debo concatenar.

```python
import pandas as pd
from gdrive import drive_list_files, download_from_drive, check_existing_file

folder_id = '1uepuL5iCVYies7GrC1YOW9Fws3NxA-Ex'
```

```python
# 2020 a 2022

file_name = 'rawson 2020-01-01 to 2022-06-30.csv'
existing_files = drive_list_files(folder_id)
    
for file in existing_files:
    if file['title'] == file_name:
        file_id = file['id']
        break

if file_id:
    file_path = download_from_drive(file_id, file_name)

    clima1 = pd.read_csv(file_path)

clima1
```

```python
# Columnas que quiero dejar
selected_columns = ['datetime', 'tempmax', 'tempmin', 'temp', 'precip', 'windspeed', 'cloudcover']

clima1 = clima1[selected_columns]

clima1
```

```python
# 2022 a 2024
file_name = 'rawson 2022-07-01 to 2024-07-31.csv'
    
for file in existing_files:
    if file['title'] == file_name:
        file_id = file['id']
        break

if file_id:
    file_path = download_from_drive(file_id, file_name)

    clima2 = pd.read_csv(file_path)

clima2 = clima2[selected_columns]

clima2
```

```python
# Concateno
df_clima = pd.concat([clima1, clima2], ignore_index=True).drop_duplicates()

# Renombro y me aseguro que la columna tenga formato fecha
df_clima.rename(columns={'datetime': 'date'}, inplace=True)

df_clima['date'] = pd.to_datetime(df_clima['date'])

df_clima
```

```python
# Guardo en Drive
existing_file_name = 'clima.csv'
check_existing_file(folder_id, existing_file_name, df_clima, 'csv')
```
