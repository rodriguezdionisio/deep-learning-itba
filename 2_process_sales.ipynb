# Process sales data

Partiendo de la raw data de ventas, productos y categorías, limpio y mergeo los datasets

```python
import pandas as pd
from gdrive import drive_list_files, download_from_drive, check_existing_file
```

## Ventas

```python
folder_id = '1uepuL5iCVYies7GrC1YOW9Fws3NxA-Ex'
file_name = 'raw_sales.csv'
existing_files = drive_list_files(folder_id)
    
for file in existing_files:
    if file['title'] == file_name:
        file_id = file['id']
        break

if file_id:
    file_path = download_from_drive(file_id, file_name)

sales_data = pd.read_csv(file_path)

sales_data
```

```python
# Renombrar columnas eliminando prefijos
sales_data.rename(columns=lambda x: x.replace('attributes.', '').replace('relationships.', ''), inplace=True)
    
sales_data = sales_data[sales_data['canceled'] != True]

# Eliminar columnas redundantes
columns_to_delete = ['id', 'type', 'canceled', 'cancellationComment', 'comment', 'price', 'status', 
                         'priceList.data', 'priceList.data.type', 'priceList.data.id', 'product.data.type', 
                         'sale.data.type', 'sale.data.id', 'subitems.data']
sales_data = sales_data.drop(columns=columns_to_delete, errors='ignore')

# Convertir columna de fecha a datetime
sales_data['createdAt'] = pd.to_datetime(sales_data['createdAt']).dt.date
```

```python
sales_data
```

## Productos

```python
folder_id = '1uepuL5iCVYies7GrC1YOW9Fws3NxA-Ex'
file_name = 'raw_products.csv'
existing_files = drive_list_files(folder_id)
    
for file in existing_files:
    if file['title'] == file_name:
        file_id = file['id']
        break

if file_id:
    file_path = download_from_drive(file_id, file_name)

products_data = pd.read_csv(file_path)

products_data
```

```python
products_data = products_data[['id', 'attributes.name', 'attributes.active', 'relationships.productCategory.data.id']]

products_data.rename(columns={'attributes.name': 'product', 
                              'attributes.active': 'active', 
                              'relationships.productCategory.data.id': 'category_id'}, inplace=True)

products_data
```

```python
df = pd.merge(sales_data, products_data, left_on='product.data.id', right_on='id', how='left')

df = df[df['active'] == True]

df.drop(['product.data.id', 'id', 'active'], axis=1, inplace=True)

df
```

## Categorías

```python
folder_id = '1uepuL5iCVYies7GrC1YOW9Fws3NxA-Ex'
file_name = 'product_categories.csv'
existing_files = drive_list_files(folder_id)
    
for file in existing_files:
    if file['title'] == file_name:
        file_id = file['id']
        break

if file_id:
    file_path = download_from_drive(file_id, file_name)

categories_data = pd.read_csv(file_path)

categories_data
```

```python
categories_data = categories_data[['id', 'attributes.name']]

categories_data.rename(columns={'id': 'category_id', 
                                'attributes.name': 'category'}, inplace=True)

categories_data
```

## Dataset final

```python
df = pd.merge(df, categories_data, on='category_id', how='left')

df.drop(['category_id'], axis=1, inplace=True)

df
```

```python
# Guardo en Drive
folder_id = '1uepuL5iCVYies7GrC1YOW9Fws3NxA-Ex'
existing_file_name = 'sales.csv'
check_existing_file(folder_id, existing_file_name, df, 'csv')
```
